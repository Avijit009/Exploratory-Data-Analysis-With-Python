```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("Ultra_Marathon_Dataset.csv")
```

```python
df.head(6)
```

df.dtypes

**Clean Up Data

```python
# df[df['Event distance/length'] == '50km']
# df[df['Event distance/length'] == '50k']
# df[df['Event distance/length'] == '50m']
df[df['Event distance/length'] == '50mi']
```

Combine 50km and 50mi

```python
df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020)]
```

```python
# df[df['Event name'] == "Everglades 50 Mile Ultra Run (USA)"]['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)
```

```python
# df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

Combine all filter data

```python
filtered_data = df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020) &(df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA') ]
```

```python
# filtered_data
```

Remove USA from events name

```python
# filtered_data['Event name'].str.split('(').str.get(0)
```

```python
filtered_data['Event name'] = filtered_data['Event name'].str.split('(').str.get(0)
```

```python
filtered_data.head()
```

Clean up based on age

```python
filtered_data['Athlete age'] = 2020 - filtered_data['Athlete year of birth']
```

```python
filtered_data.head(10)
```

Remove h from Athlete performance column

```python
# filtered_data['Athlete performance'].str.split(' ').str.get(0)
```

```python
filtered_data['Athlete performance'] = filtered_data['Athlete performance'].str.split(' ').str.get(0)
```

```python
# filtered_data
```

Drop column: Athlete Club, Athlete Country, Athlete Year of birth, Athlete Age category

```python
filtered_data = filtered_data.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'],axis=1)
```

```python
# filtered_data
```

Clean up null values

```python
# filtered_data[filtered_data['Athlete age'].isna()==1]
```

```python
filtered_data = filtered_data.dropna()
```

```python
filtered_data.shape
```

```python
# filtered_data[filtered_data.duplicated()==True]
```

Reset index

```python
filtered_data.reset_index(drop=True)
```

Fix types

```python
filtered_data.dtypes
```

```python
filtered_data['Athlete age'] = filtered_data['Athlete age'].astype(int)
```

```python
filtered_data['Athlete average speed'] = filtered_data['Athlete average speed'].astype(float)
```

```python
# filtered_data.dtypes
```

Rename Columns

```python
filtered_data = filtered_data.rename(columns = {'Year of event': 'year',
                                                'Event dates': 'race_day',
                                                'Event name': 'race_name',
                                                'Event distance/length': 'race_distance',
                                                'Event number of finishers': 'number_of_finishers',
                                                'Athlete performance': 'athlete_performance',
                                                'Athlete gender': 'athlete_gender',
                                                'Athlete average speed': 'athlete_average_speed',
                                                'Athlete ID': 'athlete_id',
                                                'Athlete age': 'athlete_age'
                                                })
```

```python
# filtered_data.head()
```

Reorder columns

```python
reorder_data = filtered_data[['race_day','race_name','race_distance','number_of_finishers','athlete_id','athlete_gender','athlete_age','athlete_performance','athlete_average_speed']]
```

```python
reorder_data.head()
```

Charts and graphs

Differnce in speed for male and female (50mi and 50km)

```python
reorder_data.groupby(['race_distance','athlete_gender'])['athlete_average_speed'].mean()
```

```python
mean_count = reorder_data.query('race_distance == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('count',ascending = False).query('count>19')
```

```python
mean_count.head(5)
```

```python
sns.histplot(reorder_data, x = 'race_distance', hue = 'athlete_gender')
```

```python
sns.displot(reorder_data[reorder_data['race_distance'] == '50mi']['athlete_average_speed'])
```

```python
sns.violinplot(data = reorder_data, x = 'race_distance',y = 'athlete_average_speed' , hue = 'athlete_gender', split = True, inner = 'quart')
```

```python
sns.lmplot(data = reorder_data, x = 'athlete_age',y = 'athlete_average_speed' , hue = 'athlete_gender')
```
